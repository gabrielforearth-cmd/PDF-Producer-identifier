```python
!pip install pypdf
import re
from pypdf import PdfReader

def get_pdf_version(path):
    # Lê os primeiros bytes do PDF para pegar o header %PDF-x.y
    with open(path, "rb") as f:
        head = f.read(16)
    m = re.search(br"%PDF-(\d\.\d)", head)
    return m.group(1).decode() if m else None

def get_pdf_producer(path):
    reader = PdfReader(path)
    meta = reader.metadata  # Dicionário com metadados
    if meta:
        return meta.get("/Producer") or meta.get("Producer")
    return None

# 🔹 Informe aqui o caminho do PDF
pdf_path = "/content/1e6cc754A8425000.pdf"

version = get_pdf_version(pdf_path)
producer = get_pdf_producer(pdf_path)

print(f"PDF version: {version or 'Não identificado'}")
print(f"Producer: {producer or 'Não identificado'}")
```
